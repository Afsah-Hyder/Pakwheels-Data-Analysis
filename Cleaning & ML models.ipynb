```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df=pd.read_csv('pakwheel.csv')
df
```

```python
# new column
from datetime import date

time=[]
today=str(date.today())
year1,month1,date1=today.split('-')
day1=int('13')
month1=int(month1)
year1=int(year1)

for i in range(len(df.date_added)):
    x=df.date_added[i].lower().strip('updated ').strip('about ')
    if 'hour' in x or 'minutes' in x or 'minute' in x:
        date='{}/{}/{}'.format(day1,month1,year1) 
    elif 'days' in x or 'day' in x:
        days=int(x.strip(' days ago').strip(' day ago'))
        if day1-days<=0:
            days=30-(days-day1)
            date='{}/{}/{}'.format(days,month1-1,year1)
        else:
            date='{}/{}/{}'.format(day1-days,month1,year1)
    elif 'months' in x or 'month' in x:
        months=int(x.strip(' months ago').strip(' month ago'))
        if month1-months<=0:
            months=12-(months-month1)
            date='{}/{}/{}'.format(day1,months,year1-1)
        else:
            date='{}/{}/{}'.format(day1,month1-months,year1)
    elif 'years' in x or 'year' in x:
        years=int(x.strip('updated ').strip(' years ago').strip(' year ago'))
        date='{}/{}/{}'.format(day1,month1,year1-years)
    else:
        date='error'
    time.append(date)
    
df.insert(9,'Time_added',time)
df
```

```python
cost=[]
for i in range(len(df.price)):
    x=df.price[i].lower().replace(',','')
    if 'lacs' in x:
        num=float(x.strip('pkr ').strip(' lacs'))*100000
    elif 'crore' in x:
        num=float(x.strip('pkr ').strip(' crore'))*10000000
    elif x=='call':
        num='Call'
    elif 'pkr' in x:
        num=float(x.strip('pkr '))
    else:
        num='error'
    cost.append(num)
df.insert(10,'Cost',cost)
df
```

```python
engine_cap=[]
for i in range(len(df.engine_capacity)):
    x=df.engine_capacity[i].lower()
    num=float(x.strip(' cc').strip(' kwh'))
    if 'kwh' in x:
        hp=num*1000/746
        cc=hp*22
        value=round(cc)
    else:
        value=x.strip(' cc')
    engine_cap.append(value)
df.insert(11,'Engine_cap_cc',engine_cap)
df
```

```python
df['mileage']=df['mileage'].str.replace(',','').str.replace(' km','')
df.rename(columns={'mileage':'mileage_km'},inplace=True)
df
```

```python
df=df.drop(columns=['date_added','engine_capacity','price'])
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import mean_squared_error 
from math import sqrt
```

```python
y=df2.Cost
x=df2.drop(['Cost'],axis=1)
x1_train,x1_test,y1_train,y1_test=train_test_split(x,y,test_size=0.3, random_state=np.random)
x1_train_scaled=scaler.fit_transform(x1_train)
x1_test_scaled=scaler.transform(x1_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error 
from math import sqrt
model=LinearRegression().fit(x1_train_scaled,y1_train)
y_pred=model.predict(x1_test_scaled)
r_sq=model.score(x1_test_scaled,y1_test)
r2_sq=model.score(x1_train_scaled,y1_train)
error = sqrt(mean_squared_error(y1_test,y_pred))
print('RMSE =', error)
print('training data:',r2_sq*100)
print('test data:',r_sq*100)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
for i in range(1,10,2):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(x1_train_scaled, y1_train)
    y_pred=knn.predict(x1_test_scaled)
    error = sqrt(mean_squared_error(y1_test,y_pred))
    print('Test data score for k= :',i,'is', r2_score(y1_test, y_pred)*100)
    print('RMSE =', error)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error,r2_score

lasso_model = Lasso().fit(x1_train_scaled,y1_train)
y_pred=lasso_model.predict(x1_test_scaled)
error = sqrt(mean_squared_error(y1_test,y_pred))
print('Traning data score: ',lasso_model.score(x1_train_scaled,y1_train)*100)
print('Test data score: ',lasso_model.score(x1_test_scaled,y1_test)*100)
print('RMSE =', error)
```

```python
from sklearn.ensemble import RandomForestRegressor
for i in range(2,20,2):
    model = RandomForestRegressor(n_estimators = i, random_state = 0)
    model.fit(x1_train_scaled,y1_train)
    y_pred=model.predict(x1_test_scaled)
    error = sqrt(mean_squared_error(y1_test,y_pred))
    print('Traning data score for',i,'decision trees:',model.score(x1_train_scaled,y1_train)*100)
    print('Test data score for',i,'decision trees:',model.score(x1_test_scaled,y1_test)*100)
    print('RMSE =', error)
```
